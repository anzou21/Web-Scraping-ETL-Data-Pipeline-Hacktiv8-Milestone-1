# A. Extract

## Melakukan import pada pandas, bs4 serta selenium agar dapat menggunakan pandas, BeautifulSoup dan Selenium


```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
```

*   **import pandas as pd** digunakan untuk memanggil library pandas agar dapat digunakan pada program, serta merubah nama pandas menjadi pd agar lebih mudah digunakan

*   **from bs4 import BeautifulSoup** digunakan untuk memanggil BeautifulSoup dari library bs4, agar dapat melakukan parsing dari html lebih

*   **from selenium import webdriver** digunakan untuk memanggil webdriver dari Selenium, agar dapat mengontrol chrome secara otomatis.

*   **import requests** digunakan untuk memanggil library requests, agar dapat mengirim http request kepada sebuah website agar dapat mengambil kontennya (scraping)

## Melakukan Webscraping terhadap website

```python
import time
driver = webdriver.Chrome()
nama_toko = []
nama_produk = []
harga_produk = []
jumlah_review = []
banyak_ukuran = []

for halaman in range(1,11):
    weblink = f'https://www.sephora.co.id/categories/fragrance?page={halaman}'
    driver.get(weblink)
    time.sleep(5)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')


    for kotak in soup.find_all('div', {'class':'products-card-container'}):
        try:
            nama_toko.append(kotak.find('div', {'class':'brand'}).get_text())
        except:
            nama_toko.append(None)
            
        try:
            harga_produk.append(kotak.find('span', {'class':'sell-price'}).get_text())
        except:
            harga_produk.append(None)

        try:
            nama_produk.append(kotak.find('div', {'class':'product-name'}).get_text())
        except:
            nama_produk.append(None)
        
        try:
            jumlah_review.append(kotak.find('span', {'class':'reviews-count'}).get_text())
        except:
            jumlah_review.append(None)

        try:
            banyak_ukuran.append(kotak.find('div', {'class':'variants-count'}).get_text())
        except:
            banyak_ukuran.append(None)
            
driver.close()
```



*   Import time digunakan untuk memanggil modul time, agar program bisa melakukan jeda dengan time.sleep()

*   driver = webdriver.Chrome() digunakan untuk membuka browser Google Chrome secara otomatis menggunakan Selenium WebDriver.

*   List kosong digunakan untuk menampung hasil scraping yang akan dilakukan

*   Looping for halaman digunakan untuk melakukan perulangan dari halaman 1 sampai 10 saat mengambil data pada produk

*   weblink = '  ' digunakan untuk memberikan link yang ingin dibuka

*   driver.get(weblink) digunakan untuk membuka halaman yang diinginkan

*   time.sleep(10) digunakan untuk memberikan jeda selama 10 detik per halaman agar data bisa ditarik dan tidak langsung berpindah halaman

*   html = driver.page_source digunakan untuk mengambil seluruh kode HTML dari halaman yang sedang terbuka

*   soup = BeautifulSoup(html, 'html.parser') digunakan untuk mengubah kode html agar dapat memudahkan mencari class tertentu

*   soup.find_all('div', {'class' : '  '}) digunakan untuk mencari semua kotak produk berdasarkan class tertentu

*   try...except digunakan untuk menghindari error jika data tidak ada, jika data tidak ada maka akan mengisinya dengan none, dan jika data ada maka akan mengambil teksnya menggunakan .get_text()

*   driver.close() digunakan untuk menutup browser Chrome setelah proses scraping selesai.

*   df = pd.DataFrame() digunakan untuk membuat sebuah dataframe kosong menggunakan pandas

*   df['nama kolom'] = 'nama variable' digunakan untuk mengisi setiap dataframe dengan data yang baru didapat dari scraping




## Membuat dataframe dari data yang telah diambil dari webscraping

```python
df = pd.DataFrame()
df['nama_toko'] = nama_toko
df['nama_produk'] = nama_produk
df['harga_produk'] = harga_produk
df['jumlah_review'] = jumlah_review
df['banyak_ukuran'] = banyak_ukuran

print(df)
```

*  menggunakan df = pd.DataFrame() untuk membuat sebuah dataframe dengan menggunakan pandas dan diberi nama df
*  menggunakan df['nama kolom yang diinginkan'] = 'nama variabel kolom yang sudah dibuat' untuk memasukan data kedalam dataframe

## Melakukan Eksplorasi data sederhana

```python
## menggunakan df.head() untuk memperlihatkan 5 data pertama pada dataframe
df.head()
```

```python
## menggunakan df.info() untuk melihat keterangan terkait dataframe {nama kolom, tipe data dan missing value}
df.info()
```

## Membersihkan data pada kolom harga_produk

```python
# Menghilangkan Rp dari dan tanda titik "." dari kolom harga_produk dengan menggunakan .str.replace('yang ingin dirubah', 'dirubah menjadi apa')
df['harga_produk'] = df['harga_produk'].str.replace('Rp', '')
df['harga_produk'] = df['harga_produk'].str.replace('.', '')


## menggunakan df.head(10) untuk mencetak 10 data pertama pada dataframe
df.head(10)
```

## Membersihkan data pada kolom jumlah_review

```python
# Menghilangkan \n, "(" dan ")" dari kolom jumlah_review dengan menggunakan .str.replace('yang ingin dirubah', 'dirubah menjadi apa')
df['jumlah_review'] = df['jumlah_review'].str.replace('\n', '')
df['jumlah_review'] = df['jumlah_review'].str.replace('(', '')
df['jumlah_review'] = df['jumlah_review'].str.replace(')', '')


## menggunakan df.head(10) untuk mencetak 10 data pertama pada dataframe
df.head(10)
```

## Membersihkan data pada kolom banyak_ukuran

```python
# Menghilangkan \n, sizes dan types dari kolom banyak_ukuran dengan menggunakan .str.replace('yang ingin dirubah', 'dirubah menjadi apa')
df['banyak_ukuran'] = df['banyak_ukuran'].str.replace('\n', '')
df['banyak_ukuran'] = df['banyak_ukuran'].str.replace('sizes', '')
df['banyak_ukuran'] = df['banyak_ukuran'].str.replace('types', '')


## menggunakan df.head(10) untuk mencetak 10 data pertama pada dataframe
df.head(10)
```

## Merubah tipe data pada kolom harga_produk, jumlah_review dan banyak_ukuran menjadi Integer

```python
# Merubah tipe data dari kolom harga_produk, kolom jumlah_review dan kolom banyak_ukuran menggunakan .astype("Tipe data yang diinginkan")
df['harga_produk'] = df['harga_produk'].astype("Int64")
df['jumlah_review'] = df['jumlah_review'].astype("Int64")
df['banyak_ukuran'] = df['banyak_ukuran'].astype("Int64")


## menggunakan df untuk mencetak dataframe secara keseluruhan
df
```

## Merubah missing value menjadi integer pada kolom jumlah_review dan banyak_ukuran

```python
# menggunakan .fillna( ).astype("tipe data yang diinginkan") untuk mengisi missing value dan merubah tipe datanya
df['jumlah_review'] = df['jumlah_review'].fillna(0).astype("Int64")
df['banyak_ukuran'] = df['banyak_ukuran'].fillna(1).astype("Int64")


## menggunakan df.head() untuk menampilkan dataframe secara keseluruhan
df.head()
```
